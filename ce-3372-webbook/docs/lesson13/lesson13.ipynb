```python
import matplotlib.pyplot as plt

# Example data
x = [1, 2, 3]
y = [4, 5, 6]
z = ['A', 'B', 'C']  # Labels corresponding to each (x, y) point

# Create scatter plot
plt.scatter(x, y)

# Add labels for each point
for i in range(len(x)):
    plt.text(x[i], y[i], z[i], fontsize=12, ha='right')

# Add axis labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Labels')

# Display plot
plt.show()
```
