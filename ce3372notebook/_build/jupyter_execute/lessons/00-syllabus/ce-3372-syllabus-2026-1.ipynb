## Course Objectives: 

The purpose of this class is to study the theory and application of hydraulics, learn how to use predictive tools such as charts and computer programs, and apply these tools to the analysis and design of distribution and collection and pumping systems. Preparation of reports is an equally emphasized objective of this course.

## Knowledge, Skills, Abilities (KSA) :
By the end of this course, students will be able to:

1. Read, synthesize, and clearly communicate ideas presented in current and historical technical literature.
2. Perform flow and head loss calculations in simple systems using Excel or Jupyter Notebooks.
3. Model flow, pressure, and transport in realistic water distribution systems using EPANET or comparable professional tools.
4. Develop and apply rule-based operating schedules to maintain prescribed conditions in water distribution systems.
5. Model flow and transport in stormwater and wastewater collection systems using SWMM 5.0 or similar tools.
6. Select and size engineering materials (pipes, pumps, valves, etc.) for drinking water, stormwater, and wastewater applications.
7. Prepare professional engineering reports to support the design of drinking water, stormwater, or wastewater infrastructure.
